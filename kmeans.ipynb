```python
import numpy as np
from matplotlib import pyplot as pt

from sklearn.cluster import KMeans
```

```python
arr = np.array([
    [1,2],
    [1.4,2.2],
    [1.3,1.7],    
    [5,2.3],
    [3.5,2.1],
    [4.3,2.7],
    [3,4],
    [4.2,2.9],
    [3.3,3.7],
    [3.2,3.5],
    [1.6,2.3],
    [2.6,3.3],
    [2.9,3.8],
    [4.6,3.3],
    [3.6,2.3],
    [1.9,2.1],
    [2.6,3.3],    
    [3.6,3.1],   
    [4.9,3.1],
    [4.1,3.9],
])
X = arr[:,0]
Y = arr[:,1]

arr.shape
```

```python
pt.plot(X,Y,c="red")
pt.show()
```

```python
cl = KMeans(n_clusters=3) # K = ?
cl.fit(arr)
cent = cl.cluster_centers_

pt.scatter(X,Y,c=cl.labels_,)
pt.scatter(cent[:,0],cent[:,1],c="red",)
pt.show()
```

```python
cl = KMeans(n_clusters=4) # K = ?
cl.fit(arr)
cent = cl.cluster_centers_
cen = cl.labels_
pt.scatter(X,Y,c=cen)
pt.scatter(cent[:,0],cent[:,1],c="red")
pt.show()
```

```python
cl = KMeans(n_clusters=5) # K = ?
cl.fit(arr)
cent = cl.cluster_centers_

pt.scatter(X,Y,c=cl.labels_)
pt.scatter(cent[:,0],cent[:,1],c="red")
pt.show()
```

```python
lst = []
for k in range(2,18):
    cl = KMeans(n_clusters=k) 
    cl.fit(arr)
    i = cl.inertia_
    lst.append(i)
```

```python
cl = KMeans(n_clusters=2)
cl.fit(arr)
pt.scatter(X,Y,c = 'red')
```

```python
pt.plot(range(2,18),lst,c="red",marker="*")
pt.show()
```
